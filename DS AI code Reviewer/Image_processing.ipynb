```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
path = r"D:\my_projects\Data-Science-with-GenAI-Internship---Sep-2024_Innomatics\d-s-m-10.jpeg"
# path = r'/content/d-s-m-10.jpeg'
img = (Image.open(path))
plt.imshow(img)
plt.show()
```

```python
print("Image file Name", img.filename)
print("Image Size", img.size)
print("Image Format", img.format)
print("Image Mode", img.mode)
```

```python
# gray scale
gray = img.convert('L') # L= gray scale Luminance
plt.imshow(gray, cmap='gray')
```

```python
# convert too binary scale
binary_image = img.convert("1")
plt.imshow(binary_image)
```

# Transparency is done by alpha channel not by RGB

a. Rotation

```python
rotated_image = img.rotate(45, expand=True, fillcolor=(0, 200, 255))
plt.imshow(rotated_image)
```

b. Flipping and Transpose

```python
# flip the image left to right
img_flipped = img.transpose(Image.Transpose.TRANSPOSE)
plt.imshow(img_flipped)
```

c.Cropping

```python
# Crop the area defined by the rectangle (x1, y1, x2, y2)
# img_cropped = img.crop((left, upper, right, lower))
img_cropped = img.crop((100, 0, 450, 300))

plt.imshow(img_cropped)
```

# d. Write in image

```python
from PIL import ImageDraw, ImageFont
```

```python
img = (Image.open(path))
draw = ImageDraw.Draw(img)
font_path = r"D:\My_Laptop\Downloads\beautiful-people-font\BeautifulPeoplePersonalUse-dE0g.ttf"
font = ImageFont.truetype(font_path, size=18)

text = "Learning DataScience from University"
draw.text((50, 50), text=text, fill=(0, 0, 0), font = font)

text1 = "Online Courses"
text2 = "From Youtube"
text3 = "From Innomatics"

draw.text((50, 150), text=text1, fill=(0, 0, 0), font = font)
draw.text((50, 250), text=text2, fill=(0, 0, 0), font = font)
draw.text((50, 350), text=text3, fill=(0, 0, 0), font = font)
plt.imshow(img)
```
